```python
import pandas as pd

file_path = ('CSV Laboratorio 2 AD-B.csv')
data = pd.read_csv(file_path)

data['Ingresos Totales'] = data['Precio_unitario'] * data['Cantidad_vendida']

filtro_cantidad = data[data['Cantidad_vendida'] > 40]

filtro_precios_altos = data.nlargest(10, 'Precio_unitario')

filtro_cantidad.to_csv('productos_mas_de_40_unidades.csv', index=False)
filtro_precios_altos.to_csv('10_precios_mas_altos.csv', index=False)

print("Productos con más de 40 unidades vendidas:")
print(filtro_cantidad)

print("\n10 precios unitarios más altos:")
print(filtro_precios_altos)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = ('productos_mas_de_40_unidades.csv')
data = pd.read_csv(file_path)

ventas_agrupadas = data.groupby(['Sucursal', 'Mes'])['Cantidad_vendida'].sum().unstack()

ventas_agrupadas.plot(kind='bar', figsize=(12, 6))
plt.title('Ventas agrupadas por sucursal y mes')
plt.xlabel('Sucursal')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=0)
plt.legend(title='Mes')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Precio_unitario', y='Cantidad_vendida', hue='Sucursal', palette='Set2')
plt.title('Relación entre Precio Unitario y Cantidad Vendida')
plt.xlabel('Precio Unitario')
plt.ylabel('Cantidad Vendida')
plt.grid(True)
plt.legend(title='Sucursal')
plt.tight_layout()
plt.show()
```
